```python
import requests as req
from bs4 import BeautifulSoup as bs
```

## 이번 예제에서 알아가야 할 내용!
1. 응답이 200으로 올바르게 완료가 됐는데 수집이 안되는 경우 해결책
2. 선택자를 한 번에 가지고 오는 방법 => 주의점 있음

```python
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver")
```

```python
soup = bs(res.text, "lxml")
```

```python
# 선택자를 한 번에 복사해오는 방법
# 사이트 -> f12 -> 개발자도구 -> 요소 -> 마우스우클릭 -> copy -> copy selector
soup.select("body > div > table > tbody > tr > td.tit > a")
```

# iframe 데이터 조회 방법
- 선택자, res[200]으로 아무런 문제가 없는 경우 반드시 iframe을 의심해라
- iframe = 다른 서버에 저장된 데이터를 불러다가 보여만 주는 역할
- 원본데이터가 저장된 주소를 찾아야한다.
- src 안에 경로가 상대경로인 경우는 반드시 절대경로로 변경해야한다!
    - 사이트 주소창(절대경로)과 내가 찾고 싶은 src경로(상대경로)를 합쳐서 넣기
    - https://finance.naver.com/marketindex/ (절대경로) 
    - /marketindex/exchangeList.naver (상대경로)
    - /marketindex/ 겹치는 이 부분을 빼고 합쳐주면 된다.
    - 상대경로는 내가 보고 싶은 페이지가 바로 보이지 않음(절대경로랑 합쳐줘야함)

# 크롤링 할 때 사용하는 선택자는 4가지 뿐이다.
1. 태그 선택자 -> 태그 이름을 작성하는 것 ( h1, a ) => 복수
2. 클래스 선택자 -> 클래스 값을 작성하는 것 ( .abc, .title ) => 복수
3. 아이디 선택자 -> 아이디 값을 작성하는 것 ( #id ) => 단수(한페이지에 하나)
4. 계층 선택자(자식) -> " > "를 통해서 활용하는 것
     - nth-child(n) = 몇 번째 자식 => 단수
